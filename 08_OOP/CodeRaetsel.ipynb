```python
n = 5
steps = 0
while n != 1:
    if n % 2 == 0:
        n = n // 2
    else:
        n = n * 3 + 1
    print(n)
    steps += 1
print(steps)
```

```python
class Basis:
    def rechne(self, x):
        return x * 2
    def hallo(self):
        return "Hi"

class MysteryE(Basis):
    def rechne(self, x):
        altes_ergebnis = super().rechne(x)
        return altes_ergebnis + 10

myMystery = MysteryE()
print(myMystery.rechne(5))
print(myMystery.hallo())
```

```python
class Person:
    def __init__(self, name):
        self.n = name
    def begruessen(self, anderer):
        print(self.n + " winkt " + anderer.n)
        
class Team:
    def __init__(self):
        self.mitglieder = []
    def aufnehmen(self, person):
        self.mitglieder.append(person)
    def aufloesen(self):
        self.mitglieder.clear()
# erstelle ein team aus drei personen:
p1 = Person("Peter")
p2 = Person("Paul")
p3 = Person("Mary")
team = Team()
team.aufnehmen(p1)
team.aufnehmen(p2)
team.aufnehmen(p3)
# ... und lass sie sich gegenseitig begrüßen
for p in team.mitglieder:
    p.begruessen(p1)
```

```python
class Herz:
    def __init__(self, name):
        self.n = name
    def schlagen(self):
        print(self.n + " schlägt")

class Hirn:
    def __init__(self, name):
        self.n = name
    def denken(self):
        print(self.n + " denkt")

class Person:
    def __init__(self, name):
        self.n = name
        self.h = Herz(name)
        self.b = Hirn(name)
```

```python
text = "A<b>C</b>D"
ergebnis = ""
zustand = 0  # 0 = Text aufnehmen, 1 = Tag ignorieren

for zeichen in text:
    if zeichen == '<':
        zustand = 1
    elif zeichen == '>':
        zustand = 0
    else:
        if zustand == 0:
            ergebnis += zeichen

print(ergebnis)
```
