```python
class Teilnehmer:
    anzahl = 0
    def __init__(self, name):
        self.name = name
        Teilnehmer.anzahl += 1

anna = Teilnehmer("Anna")
ben = Teilnehmer("Ben")
clara = Teilnehmer("Clara")
print(Teilnehmer.anzahl)
```

```python
class Spardose:
    def __init__(self):
        self.inhalt = 0
    def einwerfen(self, betrag):
        self.inhalt += betrag
    # getter-methode implementieren:
    def get_inhalt(self):
        return self.inhalt

a = Spardose()
b = Spardose()

a.einwerfen(50)
a.einwerfen(30)
b.einwerfen(10)

# ersetze print(a.inhalt, b.inhalt) mit getter:
print(a.get_inhalt(), b.get_inhalt())
```

```python
class Tier:
    def __init__(self):
        self.lebendig = True  # Gemeinsames Attribut aller Tiere

    def spricht(self):
        return "..."

    def getLebendig():
        return self.lebendig

class Katze(Tier):
    def __init__(self):
        super().__init__()  # Ruft Tier.__init__ auf → setzt self.lebendig = True

    def spricht(self):
        return "Miau"

class Hund(Tier):
    def __init__(self):
        super().__init__()  # Ruft Tier.__init__ auf → setzt self.lebendig = True

    def spricht(self):
        return "Wuff"

# Beispielverwendung
tiere = [Katze(), Hund(), Tier(), Katze()]

# Gib sowohl den Laut als auch den "lebendig"-Status aus
for t in tiere:
    print(f"{t.__class__.__name__}: {t.spricht()} (lebendig: {t.lebendig})")
```
