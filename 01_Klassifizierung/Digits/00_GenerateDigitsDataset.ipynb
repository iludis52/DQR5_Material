```python
### Schritt 1: CSV erzeugen ###
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits

# Load digits dataset
digits = load_digits()
X = digits.data
y = digits.target

# Create DataFrame with feature columns and class column
n_samples, n_features = X.shape
columns = [f'feature{i}' for i in range(n_features)]
df = pd.DataFrame(X, columns=columns)
df['class'] = y

# Save to CSV
df.to_csv('Datasets/digits.csv', index=False)
```

```python
### Schritt 2: Testbilder aus dem digits-Dataset abspeichern ###
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt

# load digits dataset
digits = datasets.load_digits()

idx = np.random.permutation(len(digits.images))[:5] # Datensätze mischen, 5 ersten auswählen

for i in idx:
    plt.imsave(f"Datasets/digit_{i}_label_{digits.target[i]}.png",
               digits.images[i] / 16.0,   # auf 0–1 normieren
               cmap="gray")
```

```python
### Schritt 3: Testbilder anzeigen lassen. ###
# select one row from the dataset
row = df.iloc[101]

# select the class of the row
class_label = row['class']

# select the features of the row
features = row.drop('class')

# print the class and features of the row
print("Class:", class_label)

# reshape the features to a 8x8 2D array
image = features.values.reshape(8, 8)
print(image)

#convert the image to a PIL image and display it in the notebook
import matplotlib.pyplot as plt

# define image size
plt.figure(figsize=(2, 2))
plt.imshow(image,cmap='gray')
plt.colorbar()
plt.show()
```
