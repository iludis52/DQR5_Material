```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering

# 1. Daten generieren: Zwei ineinander verschränkte Halbmonde
# noise=0.05 sorgt für leichte Streuung, damit es realistischer ist
X, y = make_moons(n_samples=300, noise=0.05, random_state=0)

# 2. Die vier Linkage-Methoden definieren
linkage_methods = ['single', 'average', 'complete', 'ward']

# Plot-Einstellungen (1 Reihe, 4 Spalten)
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
fig.suptitle('Vergleich der Hierarchical Clustering Linkage-Methoden', fontsize=16)

# 3. Schleife über alle Methoden
for i, method in enumerate(linkage_methods):
    # Cluster-Modell erstellen (wir suchen 2 Cluster)
    model = AgglomerativeClustering(n_clusters=2, linkage=method)
    labels = model.fit_predict(X)
    
    # Plotten
    axes[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=40)
    axes[i].set_title(f"Linkage: {method.capitalize()}", fontsize=14, fontweight='bold')
    axes[i].set_xticks([]) # Achsenbeschriftung entfernen für sauberen Look
    axes[i].set_yticks([])

plt.tight_layout()
plt.subplots_adjust(top=0.85) # Platz für den Haupttitel lassen
plt.show()
```
