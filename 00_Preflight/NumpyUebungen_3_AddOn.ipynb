```python
import numpy as np
a = np.zeros((8))
print(a)

# die letzten vier elemente sollen den wert '1' besitzen
# start | stop | step
a[4:8:1]= 1
print(a)
```

```python
b = np.zeros((8))

# jedes zweite element soll den wert '1' besitzen
# start | stop | step
b[0:8:2]= 1
print(b)
```

```python
c = np.zeros((4,3))
print(c)
print("\n")
# die erste spalte soll den wert '1' erhalten, 
# die zweite und dritte spalte bleiben unverändert

c[0:4,0:1]=1
print(c)
```

```python
c = np.zeros((4,3))
# in den ersten beiden zeilen sollen die erste wert '1' sein.

c[0:2,0:1]=1
print(c)

import matplotlib.pyplot as plt
plt.figure(figsize=(2,2))
plt.imshow(c)
```

```python
# Aufgabe: erzeuge ein 4x4 Schachbrett-Muster.
d = np.zeros((4,4))

d[0:4:2,0:4:2]=1
d[1:4:2,1:4:2]=1
print(d)

import matplotlib.pyplot as plt
plt.figure(figsize=(2,2))
plt.imshow(d)
```
