```python
import numpy as np

# plotting-library von python
# z.b. scatterplots, boxplots, lineplots, barplots, histograms
import matplotlib.pyplot as plt

# erstelle ein numpy-array 8x8
z = np.zeros((7,8))

z[::2,::2] = 1
z[1::2,1::2] = 1
print(z)

# cmap = color map
plt.imshow(z, cmap='plasma')
plt.colorbar()
plt.show()
```

```python
import numpy as np

# plotting-library von python
# z.b. scatterplots, boxplots, lineplots, barplots, histograms
import matplotlib.pyplot as plt

# erstelle ein numpy-array 8x8
z = np.zeros((8,8))

# kurzform für z[0:4:1,0:4:1] = 1
z[:4,:4] = 1
# kurzform für z[4:8:1,4:8:1] = 1
z[4:,4:] = 1
print(z)

# cmap = color map
plt.imshow(z, cmap='plasma')
plt.colorbar()
plt.show()
```

```python
import numpy as np

# plotting-library von python
# z.b. scatterplots, boxplots, lineplots, barplots, histograms
import matplotlib.pyplot as plt

# erstelle ein numpy-array 8x8
z = np.zeros((8,8))

# zeile 0 und zeile 7 werden auf 1 gesetzt
z[0:8:7, 0:] = 1
# spalte 0 und spalte 7 werden auf 1 gesetzt
z[0:8, 0:8:7] = 1 
print(z)

# cmap = color map
plt.imshow(z, cmap='plasma')
plt.colorbar()
plt.show()
```
