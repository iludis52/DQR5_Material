```python
# numpy ist die Bibliothek für numerische Berechnungen in Python
# numpy wird oft mit dem Kürzel np importiert (das ist eine Konvention)
import numpy as np

# einfachste, direkte methode, um ein numpy-array zu erstellen
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a)
print(b)

# maximaler wert von a und maximaler wert von b:
print(np.max(a))
print(np.max(b))

# hier benötigt man keinen loop, weil numpy das für uns macht
# das array besteht aus floats
c = np.linspace(0,20,11)
print(c)

d = np.linspace(0,20,21)
print(d)

# das ganze nochmal mit arange:
# das array besteht aus integern
e = np.arange(0,20,2)
print(e)

# erstellt ein arange mit floats:
f = np.arange(0,20,2.0)
print(f)

# erzeugt bitte folgendes array: 10., 12.5, 15, 17.5, 20.
g = np.arange(10,20.1,2.5)
print(g)

# erzeugt bitte folgendes array: 10., 12.5, 15, 17.5, 20.
h = np.linspace(10,20,5)
print(h)
```

```python
# teil 2: slicing!

import numpy as np

# erzeuge ein array mit 8 nullen
a = np.zeros(8)
print(a)

# setze die ersten vier zahlen auf '1' und lasse die letzten vier zahlen au '0'
# start : stop : step lautet hier die syntax
a[0:4:1]=1
print(a)

# abkürzen, weil man immer bei 0 startet und immer mit 1 schreitet
a[:4]=2
print(a)

# setze in diesem array jedes zweite element auf die zahl '9'
a[0:8:2]=9
# ist genau dasselbe wie:
a[::2]=9
print(a)

a[1:8:2]=1
# ist genau dasselbe wie:
a[1::2]=1
print(a)
```

```python
# teil 3: slicing mit 2d-array!

import numpy as np

# erzeuge ein 2d array mit 5 nullen
a = np.zeros((5,5))
print(a)

###############################################################################
# ein 2d-array besitzt achsen: das sind zeilen (axis = 0) und spalten (axis = 1)
###############################################################################

# setze die erste zeile auf 1
# in python würde man schreiben: a[0][x] = ....
a[0,:]=1
print(a)

# ist dasselbe wie:
a[0,0:5]=2
print(a)

b = np.zeros((5,5))
# setze jedes erste element einer zeile auf 3
b[::,0]=3
print(b)
```

```python
import numpy as np

array = np.zeros((8, 8),dtype=int)
print("array aus nullen")
print(array,"\n")
array[::2, ::2] = 1
print(array,"\n")
array[1::2, 1::2] = 1
print(array)

# plotte mit matplotlib

import matplotlib.pyplot as plt
plt.imshow(array)
plt.colorbar()
plt.show()
```

```python
import numpy as np

array = np.zeros((8, 8),dtype=int)
array[::3,::3] = 1
array[1::3,1::3] = 2
array[2::3,2::3] = 3
print(array,"\n")

# plotte mit matplotlib

import matplotlib.pyplot as plt
plt.imshow(array)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.zeros((8, 8), dtype=int)

idx = np.arange(8)
a[idx, idx] = 1                 # Hauptdiagonale
a[idx, 8-1-idx] = 2             # Nebendiagonale

print(a, "\n")

plt.imshow(a)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 8
a = np.zeros((n, n), dtype=int)

a[0, :] = 1
a[7, :] = 1
a[:, 0] = 1
a[:, -1] = 1

print(a, "\n")

plt.imshow(a)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.zeros((20, 20), dtype=int)

a[:, ::4] = 1     # vertikale Streifen
a[::5, :] = 2     # horizontale Streifen

print(a, "\n")

plt.imshow(a)
plt.colorbar()
plt.show()
```
